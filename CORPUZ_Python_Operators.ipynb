# OPERATORS

```python
#Python Operators

print(10+9)

print(10-9)

print(10*9)

print(10/5)

print(10%5)

print(10//3)

print(10**2)
```

BITWISE OPERATORS


```python
#Bitwise Operators

print(60 & 13)

print(60 | 13)

print(60 ^ 13)
```

ASSIGNMENT OPERATORS

```python
#Assignment Operators

x=6
y=6
x+=3
y//=3

print(x) 

print(y) 
```

LOGICAL OERATORS

```python
#Logical OP

x=True
y=False
print(x and y)

print(x or y)
```

IDENTITY OPERATORS

```python
#Identity OP

print(x is y)

print(x is not y)
```
